```python
#connect to database and create a database
import mysql.connector
mydb = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'root@39',
    database = 'shop'
)

mycursor = mydb.cursor()

# mycursor.execute('create database shop')
```

```python
mycursor.execute('show databases')

for x in mycursor:
    print(x)
```

```python
# creating table inside the database 
# mycursor = mydb.cursor()

mycursor.execute('CREATE TABLE customers (name VARCHAR(225),address VARCHAR(225))')

print('Table customer created successfully')
```

```python
mycursor.execute('ALTER TABLE customers ADD COLUMN id INT AUTO_INCREMENT PRIMARY KEY')
```

```python
sql = "INSERT INTO customers (name, address) VALUES (%s, %s)"
val = ('Sajith','Trichy')
mycursor.execute(sql, val)

mydb.commit()

print(mycursor.rowcount, 'record inserted..')
```

```python
mycursor.execute('select * from customers')

myres = mycursor.fetchall()

for x in myres:
    print(x)
```
